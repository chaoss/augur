```python
import augur
from augur import logger

app = augur.Application('../augur.config.json')
gh = app.githubapi()
repolist = app.read_config('Git', 'repositories', None, [])
org = gh.api.get_organization('twitter')
for repo in org.get_repos():
    repolist.append(repo.git_url)
app.set_config('Git', 'repositories', repolist)
print(repolist)
```
